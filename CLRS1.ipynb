- 练习1.2-2

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

n = np.linspace(1, 80, 80)
algorithm1 = lambda n: 8*n**2
algorithm2 = lambda n: 64*n*np.log2(n)
plt.figure(figsize=(5,5))
plt.scatter(n, algorithm1(n), s=3, label='$8n^2$')
plt.scatter(n, algorithm2(n), s=3, label='$64n\log n$')

diffs = algorithm1(n) - algorithm2(n)
boundary = [index for index, diff in enumerate(diffs) if (diff>0) & (diffs[index-1]<0)][0]
plt.vlines(boundary, -1000, algorithm1(n)[boundary],linestyles ='--',label='n = %d' %boundary)
plt.ylim(ymin=-1000)
plt.legend()
plt.show()
```

> 在 $n\leq42$ 以内时，插入排序优于归并排序。

- 练习1.2-3

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

n = np.linspace(1, 17, 17)
algorithm1 = lambda n: 100*n**2
algorithm2 = lambda n: 2**n
plt.figure(figsize=(5,5))
plt.scatter(n, algorithm1(n), s=3, label='$8n^2$')
plt.scatter(n, algorithm2(n), s=3, label='$64n\log n$')

diffs = algorithm1(n) - algorithm2(n)
boundary = [index for index, diff in enumerate(diffs) if (diff<0) & (diffs[index-1]>0)][0]
plt.vlines(boundary, -1, algorithm1(n)[boundary],linestyles ='--',label='n = %d' %boundary, alpha = 0.5)
plt.ylim(ymin=-1)
plt.xlim(10,18)
plt.legend()
plt.show()
```

> n 最小为15开始，算法1的运行时间优于算法2.
